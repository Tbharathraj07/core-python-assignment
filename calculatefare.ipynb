```python
# Function to calculate fare for a single trip
def calculate_fare(distance):
    base_fare = 50
    distance_rate = 10
    return base_fare + (distance * distance_rate)

trips = [5, 10, 3]
total_fare=0

for i in range(len(trips)):
    distance=trips[i]
    fare=50+(distance * 10)
    print(f"Trip {i+1}: ${fare}")
    total_fare+=fare

print(f"Total Fare: ${total_fare}")
```
