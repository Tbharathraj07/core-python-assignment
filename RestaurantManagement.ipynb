```python
# Restaurant Menu Management
initial_menu = ["Pizza", "Burger", "Pasta", "Salad"]

add_item = "Tacos"
remove_item = "Salad"
check_item = "Pizza"

def add_items(menu, new_item):
    if new_item not in menu:
        menu.append(new_item)
        return menu


def remove_items(menu, item_to_remove):
    if item_to_remove in menu:
        menu.remove(item_to_remove)
    else:
        print(f"Item {item_to_remove} not found in the menu.")
    return menu

def check_items(menu, item_to_check):
    if item_to_check in menu:

        return "pizza is Available"

    else:
        print(f"{item_to_check} is not available in the menu.")
        return False

menu = add_items(initial_menu, add_item)
menu2 = remove_items(menu, remove_item)
availability = check_items(menu2, check_item)

print(f"{menu} - your menu is updated")
print("Availability:", availability)
```
